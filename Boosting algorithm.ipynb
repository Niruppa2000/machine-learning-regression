```python
import pandas as pd
import xgboost as xgb #library
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split #through sklearn train and test takes place
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
regressor=AdaBoostRegressor(n_estimators=100, learning_rate=0.1)
regressor.fit(X_train, Y_train)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
regressor=GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=4)
regressor.fit(X_train, Y_train)
```

```python
regressor=xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=4)
regressor.fit(X_train, Y_train)

```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score #r_score and adjusted r2_score takes place
r_score=r2_score(Y_test,y_pred)
```

```python
r_score 
```

```python
#Adaboost score-0.86136 , Gradient boost score--0.8671,Xgb score---0.87827
```
